```python
import random
import numpy as np
import math
```

```python
def get_val(x,y,b,t):
    ui = (3/7)*min(x%15, 14-x%15)
    vi = (3/7)*min(x//15, 14-x//15)
    uj = (3/7)*min(y%15, 14-y%15)
    vj = (3/7)*min(y//15, 14-y//15)
    b4 = 0.1*(t//10)
    b5 = 1+0.4*(t%10)
    
    b0 = 0
    b1 = 0
    b2 = 0
    b3 = 0
    
    if (b==0):
        b0 = 1+(ui+vi)/6
        b2 = 1+(uj+vj)/6
        b1 = 1+ui/3
        b3 = 1+uj/3
    if (b==1):
        b0 = 1+(ui+vi)/6
        b2 = 1+(uj+vj)/6
        b1 = 1+vi/3
        b3 = 1+vj/3
    if (b==2):
        b0 = 1+4*math.sin(math.pi*ui/12)
        b2 = 1+4*math.sin(math.pi*uj/12)
        b1 = 1+(36-(6-ui)*(6-ui))*(36-(6-vi)*(6-vi))/324
        b3 = 1+(36-(6-uj)*(6-uj))*(36-(6-vj)*(6-vj))/324
    if (b==3):
        b0 = 1+4*math.sin(math.pi*vi/12)
        b2 = 1+4*math.sin(math.pi*vj/12)
        b1 = 1+(36-(6-ui)*(6-ui))*(36-(6-vi)*(6-vi))/324
        b3 = 1+(36-(6-uj)*(6-uj))*(36-(6-vj)*(6-vj))/324
    rt = b0+b1*random.random()+b2+b3*random.random()+b4+b5*random.random()+random.random()
    return rt
    
```

```python
def get_dataset():
    ## Create an np matrix of size 625*625(spatial)*4(band)*100*temporal
    data = np.zeros((225,225,4,100))
    for x in range(0, 225):
        #print(x)
        for y in range(0,225):
            for b in range(0,4):
                for t in range(0,100):
                    data[x][y][b][t] = get_val(x,y,b,t)
    return data
    
```

```python
simulated = get_dataset()
```
